```python
import matplotlib.pyplot as plt

a=[1,2,3,4,5,6,7]
y1=[22,11,44,25,77,56,30]
y2=[22,21,54,30,70,46,40]
plt.plot(a,y1)
plt.plot(a,y2)
plt.show()
```

```python
#plot figure size of your own desire
a=[1,2,3,4,5,6,7]
y1=[22,11,44,25,77,56,30]
y2=[22,21,54,30,70,46,40]
plt.figure(figsize=[10,4])#[width,Heught]
plt.plot(a,y1)
plt.plot(a,y2)
plt.show()
```

```python

#Working with labels

a=[1,2,3,4,5,6,7]
y1=[22,11,44,25,77,56,30]
y2=[22,21,54,30,70,46,40]
plt.plot(a,y1,label='Old_Price')
plt.plot(a,y2,label='New_Price')
plt.legend(loc=2)#Inside can give various loc=int values to assign position of the denoter
#                 Loc values can go upto 10 
plt.show()
```

```python
#we can write simply in:
a=[1,2,3,4,5,6,7]
y1=[22,11,44,25,77,56,30]
y2=[22,21,54,30,70,46,40]
plt.plot(a,y1)
plt.plot(a,y2)
plt.legend()
plt.show()
```

```python
#While there are multiple values and we have to give denotations
import matplotlib.pyplot as plt
a=[1,2,3,4,5,6,7]
y1=[22,11,44,25,77,56,30]
y2=[22,21,54,30,70,46,40]
y3=[56,19,64,22,66,96,60]
y4=[67,45,41,56,50,43,33]
plt.plot(a,y1,label='a1')
plt.plot(a,y2,label='a2')
plt.plot(a,y3,label='a3')
plt.plot(a,y4,label='a4')

plt.legend(loc='upper right',bbox_to_anchor=(1,1.2),ncols=2)#bbox give location of legend  outof chart
#                ncols gives the number of columns of a legend

#we can save figure by using plt.savefig()
plt.savefig('legend.png')#while saving you can give multiple attributes to photo like facecolor and so other
plt.show()
```
