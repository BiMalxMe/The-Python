```python
from matplotlib import pyplot as plt
import pandas as pd

# Read the data from the Excel file
data = pd.read_excel('SaleData.xlsx')

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Group by 'Manager' and calculate the mean of 'Units'
groupby = df.groupby("Manager")['Units'].mean()

# Print the grouped data
print(groupby)

# Plot the results
plt.step(groupby.index, groupby.values, marker='o')
plt.show()
```
