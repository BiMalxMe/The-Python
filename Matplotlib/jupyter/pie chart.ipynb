```python
import matplotlib.pyplot as plt

Brand=['Oneplus','Samsung',"Vivo",'Google Pixel','Apple']
Ratings=[7.5,8.5,5,7,9]
ex=[0,0,0,0,0.1]
#you can also add various colors
plt.pie(Ratings,labels=Brand,explode=ex,shadow=True)#Explode highlights the certain piece in pie,,Giving shadow for 3d feel
plt.show()
```

```python
Brand=['Oneplus','Samsung',"Vivo",'Google Pixel','Apple']
Ratings=[7.5,8.5,5,7,9]
ex=[0,0,0,0,0.1]
plt.pie(Ratings,labels=Brand,explode=ex,shadow=True,autopct='%.1f',startangle=90,radius=2)#autopct writes value inside the pie chart
#statangle gives starting degree to the chart,Increasing the radius of the chart
plt.show()
```

```python
#piechart from the exisiting data
import pandas as pd

data=pd.read_excel('SaleData.xlsx')
x=data['Region']
y=data["Unit_price"]
plt.pie(y,radius=2,labels=x ,shadow=True,autopct='%.1f')
plt.show()
```

```python
datas=pd.read_excel('SaleData.xlsx')
data=pd.DataFrame(datas)

group_by=data.groupby('Region')['Unit_price'].sum()
print(group_by)
plt.pie(group_by.values,labels=group_by.index,startangle=90)
plt.show()
```
