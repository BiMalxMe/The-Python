```python
import matplotlib.pyplot as plt

Day=[1,2,3,4,5,6,7,8,9]
Sale1=[2000,3000,5000,8000,9000,11000,13000,14000,14500]
Sale2=[11000,13000,15000,18000,19000,21000,23000,24000,24500]
Sale3=[22000,23000,25000,28000,29000,31000,323000,34000,34500]
plt.stackplot(Day,Sale1,Sale2,Sale3,hatch='*',baseline='sym',colors=['hotpink','red','cyan'])#hatch fills the shapes in the plot,baseline is symmetry
plt.show()
```
