```python
import matplotlib.pyplot as plt
import seaborn as snc

data=snc.load_dataset('tips')
snc.boxplot(data,x='tip')
plt.show()


#we can see MIN,Q1,MEDIAN.Q3,MAX,OUTLIERS SIMULTANEOUSLY
```

```python
#inputing x and y 

data=snc.load_dataset('tips')
snc.boxplot(data,x='sex',y='tip',orient='vertical')#mostly put y=integer value
#it turn box inside a plot vertically erect
plt.show()
```
