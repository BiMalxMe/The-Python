```python
import matplotlib.pyplot as plt
import seaborn as sns

data=sns.load_dataset('tips')
sns.relplot(data=data,x='tip',y='total_bill',hue='sex')#same as scatter plot
plt.show()
```

```python
#for changing the type of plot
import matplotlib.pyplot as plt
import seaborn as sns

data=sns.load_dataset('tips')
sns.relplot(data=data,x='tip',y='total_bill',hue='sex',kind='line')#same as scatter plot
plt.show()
#can only change into scatter or line
```

```python
#spliting the plots on the basis of days in rows
import matplotlib.pyplot as plt
import seaborn as sns

data=sns.load_dataset('tips')
sns.relplot(data=data,x='tip',y='total_bill',hue='sex',kind='line',row='day')
plt.show()
```

```python
#spliting the plots on the basis of days in cols
import matplotlib.pyplot as plt
import seaborn as sns

data=sns.load_dataset('tips')
sns.relplot(data=data,x='tip',y='total_bill',hue='sex',kind='line',col='day')
plt.show()
```
