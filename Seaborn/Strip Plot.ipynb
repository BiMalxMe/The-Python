```python
#You can clearly see the difference
import matplotlib.pyplot as plt
import seaborn as snc

data=snc.load_dataset('tips')
snc.scatterplot(data,x="day",y="total_bill")
plt.show()
print(data)
```

```python
#Exploring more properties
import matplotlib.pyplot as plt
import seaborn as snc

data=snc.load_dataset('tips')
snc.stripplot(data,x="day",y="total_bill",hue='sex',dodge=True)
#dodge is used to seperate the values of hue,also to expand datas we can write jitt
plt.show()
print(data)
```
