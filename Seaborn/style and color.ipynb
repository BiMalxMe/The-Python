```python
import matplotlib.pyplot as plt
import seaborn as snc

data=snc.load_dataset('exercise')
print(data)
snc.barplot(data,x="time",y='pulse')
plt.show()
```

```python
#Chnging the bg color to dark
data=snc.load_dataset('exercise')
snc.set_style(style='dark')#can see other using shift tab
print(data)
snc.barplot(data,x="time",y='pulse')
plt.show()
```

```python
# Giveing other styles
data=snc.load_dataset('exercise')
snc.set_style(style='ticks')
print(data)
snc.barplot(data,x="time",y='pulse')
#ticks are "-" shown along the a and y axis labels to preacise the data
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the color palette
palette = sns.color_palette('magma', n_colors=2)

# Create a figure and axis
plt.figure(figsize=(4 ,2))
plt.imshow([palette], aspect='auto')
plt.axis('of')  # Hide the axes

# Display the palette
plt.show()
```
