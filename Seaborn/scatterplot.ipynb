```python
import matplotlib.pyplot as plt
import seaborn as snc

data=snc.load_dataset('tips')
print(data)
snc.scatterplot(data,x='total_bill',y='tip',hue='sex')
# plt.legend(loc='upper right')  #was just trying if it was working or not
plt.show()
```

```python
#using size parameter
import matplotlib.pyplot as plt
import seaborn as snc

data=snc.load_dataset('tips')
print(data)
snc.scatterplot(data,x='total_bill',y='tip',hue='sex',size='size',ncols=2)
plt.show()
```
