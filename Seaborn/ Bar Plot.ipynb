```python
import matplotlib.pyplot as plt
import seaborn as sns

# data=sns.get_dataset_names()#Helps to find out dataset from the web
data=sns.load_dataset('tips')
sns.barplot(data=data,x='day',y='tip')
plt.show()
```

```python
#To find bar of median values
data=sns.load_dataset('tips')
sns.barplot(data=data,x='day',y='tip',estimator='median',hue='sex',palette='magma')#,errorbar=('ci', 0) to remove error bar
plt.show()
```

```python
# You can give order of the x also
#for example
data=sns.load_dataset('tips')
sns.barplot(data=data,x='day',y='tip',hue='sex',order=['Sun',"Sat",'Fri','Thur',bbox_to_anchor=(1,2))#Reversed the previous order
plt.show()
```
