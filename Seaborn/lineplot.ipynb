```python
import pandas as pd
import seaborn as snc
import matplotlib.pyplot as plt
list={
    "Name":['Bimal','Kristi','Sambhav','Roshan','Girish'],
    "Marks":[67,87,76,54,86]
}
data=pd.DataFrame(list)
snc.lineplot(data=data,x='Name',y='Marks',marker='o')
plt.show()
```

```python
# Plotting from excel_file
data=pd.read_excel('SaleData.xlsx')
snc.lineplot(data=data,x="Manager", y='Units',hue='Unit_price')
plt.show()#you can later add on various postion of the denoter using legend()
```

```python
#You can pass various color palette too
# Plotting from excel_file
color=snc.color_palette('magma')#can find different color palette from the web
data=pd.read_excel('SaleData.xlsx')
snc.lineplot(data=data,x="Manager", y='Units',hue='Unit_price',palette=color)
plt.show()
```
