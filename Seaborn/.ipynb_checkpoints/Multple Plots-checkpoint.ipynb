```python
import matplotlib.pyplot as plt
import seaborn as sns

data=sns.load_dataset('tips')
print(data)
a=sns.FacetGrid(data,col='day')
a.map(sns.scatterplot,'smoker','tip')
plt.show()
```

```python
# Also you can make barplots too

import matplotlib.pyplot as plt
import seaborn as sns

data=sns.load_dataset('tips')
print(data)
a=sns.FacetGrid(data,col='day',hue='sex')
a.map(sns.barplot,'smoker','tip')            #for removing error bar errorbar=('ci', 0)
plt.show()
```

```python
#scatter plots
import matplotlib.pyplot as plt
import seaborn as sns

data=sns.load_dataset('tips')
print(data)
a=sns.FacetGrid(data,col='day',hue='sex')
a.map(sns.scatterplot,'smoker','tip')            #for removing error bar errorbar=('ci', 0)
plt.show()
```
