```python
import matplotlib.pyplot as plt
import seaborn as sns

data=sns.load_dataset('tips')
sns.swarmplot(data=data,x='day',y="total_bill")
#swarm plot is different from scatter and strip plot because it doesnt overlaps the data
plt.show()
```

```python
#we can also add more space in betn data with jitter in only strip plot
data=sns.load_dataset('tips')
sns.stripplot(data=data,x='day',y='total_bill',jitter=0.3)
#swarm plot is different from scatter and strip plot because it doesnt overlaps the data
plt.show()
```
