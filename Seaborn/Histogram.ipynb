```python
import matplotlib.pyplot as plt
import seaborn as snc

data=snc.load_dataset('tips')
snc.histplot(data,x='day',hue='sex',kde=True)#hue further classifies data into simpler terms
#kde gives hist plots in a curve
plt.show()
```

```python
#Taking different datset
import matplotlib.pyplot as plt
import seaborn as snc

data=snc.load_dataset('titanic')
print(data)
snc.histplot( data,x='age',hue='who',kde=True)#,bins=5)You can give bins five too
plt.show()
```

```python
#we can make data closely paked too
import matplotlib.pyplot as plt
import seaborn as snc

data=snc.load_dataset('titanic')
snc.histplot(data,x='age',hue='who',discrete=True,kde=True)#dicrete makes closely packed
plt.show()
```
